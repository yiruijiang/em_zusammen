```python
%matplotlib inline

import matplotlib.pyplot as plt
import nest_asyncio
import pandas as pd

nest_asyncio.apply()
import sys
import os

# Add project root (adjust '..' as needed)
sys.path.append(os.path.abspath(".."))

from em.source import tools
```

```python
df = tools.retrieve_historical_stock_data(
    stock='AAPL',
    currency='USD',
    duration='5 Y'
)
plt.figure(figsize=(12, 6))
ax = plt.gca()
df.plot(x='date', y='close', title='AAPL 5-Year Close Price', ax=ax)
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```
